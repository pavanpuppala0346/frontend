#### Static Web Applications
- A static website is one with stable content, where every user sees the exact same thing on each individual page.

#### Dynamic Web Applications
- A dynamic website is one where content is pulled on-the-fly, allowing its content to change with the user.
- Ex: Instagram, Youtube, Live score updates, Latest Stats, Live traffic updates, Form data Validations
    - Similarly Gmail sign in (user interactivity)- form data validations

#### How can we build this Dynamic web applications?
- Client Server communication
- Manipulating the HTML and CSS
- Writing application logic

#### Ways to achieve
- Using JavaScript
- Using Web Assembly

#### Why JavaScript?
- it is most popular programming language which is adopted and understood by all major browsers
- Java script can be used in both Client side and Server side.
    -	We can add animations
    -	We can add live dashboards
    -	We can add browser based games
- To have dynamic experience on our website we use java script

- HTML :-  Provides the basic structure for webpages
- CSS :- it is used for styling
- JS :- It is used primarily by web browsers to create a dynamic and interactive experiences for the users.

#### 1. JavaScript Variables
#### 1.1 Variable Declaration : Variables are like containers for storing values. We can create a variable using the let keyword.
- let message;

#### 1.2 Assigning a Value to the Variable
- We can put data into a variable using an assignment operator (=).
- let message = 'Hello Pavan';
    - (or)
- let message;
- message = 'Hello Pavan';
